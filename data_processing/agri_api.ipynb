## 單一作物補齊（蘿蔔SA、胡瓜FC、甘藍LA）

```python
veg = "甘藍"
code = "LA"

# market_list = ["台北一","台北二","西螺鎮","屏東市"]
market_list = ["台中市","高雄市"]
URL = "https://data.coa.gov.tw/api/v1/AgriProductsTransType"
for market in market_list:
    df = pd.DataFrame(columns=["CropCode","CropName","MarketName","TransDate","MarketCode","Upper_Price","Middle_Price","Lower_Price","Avg_Price","Trans_Quantity"])
    print("----------------------------------")
    print(market)
    page = 1
    payload = {'Start_time':'107.01.01','End_time':'110.12.31', \
    'CropName': veg, 'MarketName':market,'api_key':'2D5BOFKM0A6LDJKER4R0N8QMIMC4M7'}
    r = requests.get(URL,params=payload).json()
    print(r)
    for res in r['Data']:
        if res['CropCode'].find(code) == -1 and res['CropCode']!="-":
            continue
        else:
            df = df.append(res,ignore_index=True)
    while r["Next"]==True:
        page += 1
#         print(page)
        payload = {'Start_time':'107.01.01','End_time':'110.12.31',\
        'CropName': veg, \
        # "CropCode": "LO1", \
        'MarketName':market,'api_key':'2D5BOFKM0A6LDJKER4R0N8QMIMC4M7','Page':page}
        r = requests.get(URL,params=payload).json()
        # print(len(r['Data']))
        for res in r['Data']:
            if res['CropCode'].find(code) == -1 and res['CropCode']!="-":
                continue
            else:
                df = df.append(res,ignore_index=True)
    print(len(df))
    
    filename = veg+"_"+market+".csv"
    df.to_csv(DIR_PATH+"/"+filename)
```

## 單一作物補齊（蘿蔔SA、胡瓜FC、甘藍LA）

```python
veg = "甘藍"
code = "LA"
# market_list = ["台北一","台北二","西螺鎮","屏東市"]
market_list = ["台中市","高雄市"]
year_list = ["106","107","108","109","110"]
month_list = [".01",".02",".03",".04",".05",".06",".07",".08",".09",".10",".11",".12"]
URL = "https://data.coa.gov.tw/Service/OpenData/FromM/FarmTransData.aspx?"

for market in market_list:
    df = pd.DataFrame(columns=["交易日期","作物代號","作物名稱","市場代號","市場名稱","上價","中價","下價","平均價","交易量"])
    print("----------------------------------")
    print(market)
    # page = 1
    for year in year_list:
        for month in month_list:
            payload = {'StartDate':year+month+'.01','EndDate':year+month+'.31','top':9999,'skip':0,'Crop':veg,'Market':market}
            r = requests.get(URL,params=payload).json()
            for res in r:
                if res['作物代號'].find(code) == -1 and res['作物代號']!="rest":
                        continue
                else:
                    df = df.append(res,ignore_index=True)
    print(len(df))
    df = df.sort_values(by='交易日期')
    filename = veg+"_"+market+".csv"
    df.to_csv(DIR_PATH2+"/"+filename)
```
